```python
import pandas as pd
```

```python
df = pd.read_csv('data/Ticket/Ticket_textual.csv', sep=';')
```

```python
df.head()
```

```python
subs, subs_inv = {}, {}
for z in df.select_dtypes(include=['object']).columns:
    subs[z] = {y:x for x, y in enumerate(df[z].unique())}
    subs_inv[z] = {x: y for x, y in enumerate(df[z].unique())}
```

```python
df2 = df.replace(subs)
df2.head()
```

```python
df2.to_csv('data/Ticket/Ticket.csv', sep=';', index=False)
```

```python
a = pd.read_csv('data/Ticket/Ticket_labels.csv', sep=';')
```

```python
len(a.iloc[:, 0].unique())
```

```python
subs_label, subs_inv_label = {}, {}
for z in a.columns:
    subs_label[z] = {y:x for x, y in enumerate(df[z].unique().astype('int'))}
    subs_inv_label[z] = {x: y for x, y in enumerate(df[z].unique().astype('int'))}
```

```python
subs_label
```

```python
subs_label.get(a.iloc[:, 0].astype('int').values[0])
```

```python
b = a.copy()
column = a.columns.tolist()[0]
b['labels'] = [subs_label[column].get(x) for x in a[column].astype('int').values]
```

```python
b
```

```python
b[['labels']].to_csv('data/Ticket/Ticket_labels.csv', index=False)
```
